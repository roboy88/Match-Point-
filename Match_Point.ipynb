```python
# Step 1: Install MoviePy
!pip install moviepy

# Step 2: Upload your files (if needed)
from google.colab import files
uploaded = files.upload()

# Step 3: Import MoviePy tools
from moviepy.editor import *

# Step 4: List of slideshow images (2 seconds each)
image_files = [
    "ChatGPT Image May 21, 2025, 01_11_22 PM.png",
    "25c34cba-2bf0-4dc3-a8f3-1f5ba0526c9e.png",
    "a9eecec2-44c7-49f3-969f-db9a5d7ba9ac.png",
    "c17b4b2f-e79b-492b-88be-f899a54aa5fb.png",
    "f78dbbae-17f9-49fc-a2c3-ac3413e3d2b2.png"
]

# Step 5: Create image clips with fade transitions
image_clips = []
for img in image_files:
    clip = ImageClip(img).set_duration(2).fadein(0.5).fadeout(0.5).resize(height=720)
    image_clips.append(clip)

# Step 6: Load video (trim to first 5 seconds, resize to match)
match_clip = VideoFileClip("Tennis clip.MOV").subclip(0, 5).resize(height=720)

# Step 7: Combine slideshow + video
final_clip = concatenate_videoclips(image_clips + [match_clip])

# Step 8: Export final video
final_clip.write_videofile("match_point_trailer.mp4", fps=24)
```
